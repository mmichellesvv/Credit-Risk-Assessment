3. do binning of numerical data before WoE-preprocessing;

```python
def bin_continuous(series, bins=20, bin_edges=None, method="qcut"):
    
    series = series.copy()
    nan_mask = series.isna()

    if bin_edges is not None:
        binned = pd.cut(series[~nan_mask], bins=bin_edges, include_lowest=True, duplicates="drop")
    else:
        if method == "qcut":
            binned, bin_edges = pd.qcut(series[~nan_mask], q=bins, retbins=True, duplicates="drop")
        else:
            min_val, max_val = series.min(), series.max()
            bin_edges = np.linspace(min_val, max_val, bins + 1)
            binned = pd.cut(series[~nan_mask], bins=bin_edges, include_lowest=True, duplicates="drop")

    binned = binned.astype(str)
    binned[nan_mask] = "Missing"

    return binned, bin_edges
```

4. do WoE-preprocessing;

```python
def calculate_woe_iv(feature, binned_series, target_series, eps=1e-6):
    
    temp = pd.DataFrame({'feature': feature, 'bin': binned_series, 'target': target_series})

    grouped = temp.groupby('bin', dropna=False)['target'].agg(
        n_obs='count',
        n_bad=lambda x: (x == 0).sum(),
        n_good=lambda x: (x == 1).sum()
    ).reset_index()

    grouped['feature'] = feature
    grouped['prop_n_good'] = (grouped['n_good'] + eps) / (grouped['n_good'].sum() + eps)
    grouped['prop_n_bad']  = (grouped['n_bad']  + eps) / (grouped['n_bad'].sum()  + eps)

    grouped['woe'] = np.log(grouped['prop_n_good'] / grouped['prop_n_bad'])
    grouped['iv']  = (grouped['prop_n_good'] - grouped['prop_n_bad']) * grouped['woe']

    iv = grouped['iv'].sum()

    return grouped, iv
```

```python
def create_woe_features(X_train, X_test, y_train, num_data, cat_data, bins=20):
    
    X_train_woe = X_train.copy()
    X_test_woe = X_test.copy()
    iv_dict = {}
    df_to_save = pd.DataFrame()

    def process_feature(feature, is_numeric=True):
        
        # ===================== NUMERIC =====================
        if is_numeric:
            X_train_woe[feature], bin_edges = bin_continuous(X_train_woe[feature], bins=bins)
            X_test_woe[feature], _ = bin_continuous(X_test_woe[feature], bin_edges=bin_edges)
        else:
        # ===================== CATEGORICAL =====================
            X_train_woe[feature] = X_train_woe[feature].fillna("Missing").astype(str)
            X_test_woe[feature] = X_test_woe[feature].fillna("Missing").astype(str)

        woe_table, iv_value = calculate_woe_iv(feature, X_train_woe[feature], y_train)

        if 0.02 < iv_value < 0.5:
            
            lower = woe_table['woe'].quantile(0.05)
            upper = woe_table['woe'].quantile(0.95)
            woe_table['woe'] = woe_table['woe'].clip(lower, upper)

        # --------------------- CATCH-ALL -----------------------
            if 'Unknown' not in woe_table['bin'].values:
                mean_woe = woe_table['woe'].mean()
                woe_table = pd.concat([woe_table,
                                       pd.DataFrame({'feature':[feature],'bin':['Unknown'],'woe':[mean_woe],'iv':[0]})],
                                       ignore_index=True)

            woe_map = dict(zip(woe_table['bin'], woe_table['woe']))

            X_train_woe[feature] = X_train_woe[feature].map(woe_map)
            X_test_woe[feature] = X_test_woe[feature].map(woe_map)
            
            X_test_woe[feature] = X_test_woe[feature].fillna(woe_table['woe'].mean())

            iv_dict[feature] = iv_value
            return woe_table[['feature', 'bin', 'woe', 'iv']]
        else:
            X_train_woe.drop(feature, axis=1, inplace=True)
            X_test_woe.drop(feature, axis=1, inplace=True)
            return pd.DataFrame()

    for feature in num_data:
        df_to_save = pd.concat([df_to_save, process_feature(feature, is_numeric=True)], ignore_index=True)

    for feature in cat_data:
        df_to_save = pd.concat([df_to_save, process_feature(feature, is_numeric=False)], ignore_index=True)

    return X_train_woe, X_test_woe, iv_dict, df_to_save
```

```python
import numpy as np

X_train_woe, X_test_woe, iv_dict, df_to_save = create_woe_features(X_train, X_test, y_train,
                                                       num_data=num_data,
                                                       cat_data=cat_data,
                                                       bins=20)
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(12, 6))
bars = plt.bar(range(slsted_data.shape[0]), slsted_data['iv'], color='blue')

for i, val in enumerate(slsted_data['iv']):
    plt.text(i, val + 0.005, round(val, 3), ha='center', va='bottom', fontsize=10)

plt.xticks(range(slsted_data.shape[0]), slsted_data['feature'], rotation=45, ha='right')
plt.ylabel('IV score')
plt.title('Top features by IV score')
plt.grid(axis='y')

plt.show()
```
