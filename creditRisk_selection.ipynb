3. binning numerical data before WoE-preprocessing;

```python
def bin_continuous(series, bins=20, bin_edges=None, method="qcut"):
    """
    Binning continuous variable for WoE.
    method: "qcut" = by quantiles, "cut" = equal width
    """
    series = series.copy()
    nan_mask = series.isna()
    
    if bin_edges is not None:
        binned = pd.cut(series[~nan_mask], bins=bin_edges, include_lowest=True, duplicates="drop")
    else:
        if method == "qcut":
            binned, bin_edges = pd.qcut(series[~nan_mask], q=bins, retbins=True, duplicates="drop")
        else:
            min_val, max_val = series.min(), series.max()
            bin_edges = np.linspace(min_val, max_val, bins + 1)
            binned = pd.cut(series, bins=bin_edges, include_lowest=True, duplicates="drop")
    
    binned = binned.astype(str)
    binned[nan_mask] = "Missing"

    
    return binned, bin_edges
```

```python
import numpy as np

for feature in num_data:
    X_train[feature], bin_edges = bin_continuous(X_train[feature])
    X_test[feature], _ = bin_continuous(X_test[feature], bin_edges=bin_edges)
```

4. WoE-preprocessing whole dataset;

```python
import numpy as np

def calculate_woe_iv(binned_series, target_series, eps=1e-6):
    """
    Universal WoE & IV calculation for both continuous and discrete variables.
    """
    temp = pd.DataFrame({'feature': binned_series, 'target': target_series})

    grouped = temp.groupby('feature', dropna=False).agg(
        n_obs=('target', 'count'),
        n_good=('target', 'sum')
    ).reset_index()

    grouped['n_bad'] = grouped['n_obs'] - grouped['n_good']
    grouped['prop_n_good'] = grouped['n_good'] / grouped['n_good'].sum()
    grouped['prop_n_bad'] = grouped['n_bad'] / grouped['n_bad'].sum()

    grouped['WoE'] = np.log((grouped['prop_n_good'] + eps) / (grouped['prop_n_bad'] + eps))
    grouped['IV'] = ((grouped['prop_n_good'] - grouped['prop_n_bad']) * grouped['WoE']).sum()

    return grouped
```

```python
iv_dict = {}

for feature in num_data + cat_data:
    woe_table_train = calculate_woe_iv(X_train[feature], y_train)
    iv_value = woe_table_train['IV'].iloc[0]
    
    if 0 < iv_value < .5:
        woe_map = dict(zip(woe_table_train['feature'], woe_table_train['WoE']))
        X_train[feature] = X_train[feature].map(woe_map)
        X_test[feature] = X_test[feature].map(woe_map)  # on test we use WoE from train
        
        iv_dict[feature] = iv_value
    else:
        X_train.drop(feature, axis=1, inplace=True), X_test.drop(feature, axis=1, inplace=True)
```
